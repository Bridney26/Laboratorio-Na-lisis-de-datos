```python
#Laboratorio
from folium.plugins import MiniMap
from branca.element import Figure

poputext = '<b>Heredia</b>'

my_map = folium.Map(location=[10.2856822,-84.9008912],zoom_start=16,titles='stamenterrain')
                    
folium.Marker(location=[10.2856822,-84.9008912],popup=poputext,icon=folium.Icon(color="orange", icon="ok-sign")).add_to(my_map)    

folium.Circle(location=[10.2856822,-84.9008912],color="purple",weight=6,fill_color="red",fill_opacity=0.5,
              tooltip="Heredia").add_to(my_map)

minimap = MiniMap(tile_layer='stamenterrain', position='bottomright')

my_map.add_child(minimap)


my_map.save('C:folium.html')

my_map
```
